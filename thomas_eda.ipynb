```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
df.head
```

```python
pd.set_option('display.max_columns', None)
df.describe()
```

```python
df['years_played_NBA'] = df['active_to']-df['active_from']
df.describe()
```

```python
sns.histplot(df['years_played_NBA'], bins=15)
```

```python
sns.lmplot(
    x='NCAA_games',
    y='NBA_g_played',
    data=df,
    fit_reg=True,
    scatter_kws={'alpha':0.3}
)
```

```python
sns.lmplot(
    x='NCAA_ppg',
    y='NBA_ppg',
    data=df,
    fit_reg=True,
    scatter_kws={'alpha':0.3}
)
```

```python
sns.lmplot(
    x='NCAA__3ptpct',
    y='NBA_g_played',
    data=df,
    fit_reg=True,
    scatter_kws={'alpha':0.3}
)
```

```python
sns.lmplot(
    x='weight',
    y='NBA_g_played',
    data=df,
    fit_reg=True,
    scatter_kws={'alpha':0.3}
)
```
